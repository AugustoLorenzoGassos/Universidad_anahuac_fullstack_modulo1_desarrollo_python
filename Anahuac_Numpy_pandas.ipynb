## **Indexado de un dataframe**
Para extraer datos de un DataFrame de pandas podemos usar:
* **indexación directa**: primero columnas, luego filas
* **accesores**: primero filas, luego columnas


**Accesores**

La diferencia entre los accesores y la indexacion directa, es que con los accesores
primero se indican las filas y luego las columnas.


Los accesores  son: **.iloc** y **.loc**.

La diferencia entre ellos es que .iloc acepta índices y .loc acepta etiquetas


df.loc['rowname', 'colname']

df.iloc['rowindex', 'colindex']
