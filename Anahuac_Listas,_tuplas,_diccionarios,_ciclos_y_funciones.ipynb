**7. Dividir lista**

Divide una lista en dos listas: una con números pares y otra con impares.

```python
ListaNumeros = [200,30,13,154,75,21,35,10]

ListaPares = []
ListaNoPares = []

print("Lista original: ", ListaNumeros)

for i in ListaNumeros:
  if i%2 == 0:
    ListaPares.append(i)
  else:
    ListaNoPares.append(i)

print("Números pares: ", ListaPares)
print("Números no pares: ", ListaNoPares)
```

**14. Contar letras**

```python
Texto = "abbcccddddeeeeeffffff"
LetrasContadas = ""

Diccionario = {}

for i in range(0,len(Texto),1):
  if i == 0:
    LetrasContadas = LetrasContadas + Texto[i]
    Diccionario[Texto[i]] = Texto.count(Texto[i])
  else:
    if LetrasContadas.count(Texto[i]) == 0:
      LetrasContadas = LetrasContadas + Texto[i]
      Diccionario[Texto[i]] = Texto.count(Texto[i])

print(Diccionario)
```

**16. Agenda telefónica**

Crea tres funciones que manipule un diccionario:

* Permita agregar contactos

* Buscar un contacto

* Mostrar todos los contactos

```python
Opcion = 0
Contactos = {"Nombres":["Augusto","Juan","Mónica"],"Telefono":["5511281636","2281123456","2288123456"]}

def AgregarContacto():
  print("Agregar contacto")
  NombreContacto = input("Escribe el número de contactro: ")
  TelefonoContacto = input("Escribe el número de teléfono del contacto")
  Contactos["Nombres"].append(NombreContacto)
  Contactos["Telefono"].append(TelefonoContacto)
  print(Contactos)

def BuscarContacto():
  print("Buscar contacto")
  ItemEncontrado = False
  NombreContacto = input("Escribe el nombre del contacto a buscar: ")

  for i in range(1,len(Contactos["Nombres"])+1,1):
    if  NombreContacto == Contactos["Nombres"][i-1]:
      ItemEncontrado = True
      break

  if ItemEncontrado == True:
    print(f"Contacto {NombreContacto} encontrado con número de teléfono {Contactos["Telefono"][i-1]}")
  else:
    print(f"Contacto {NombreContacto} no encontrado")

def ListaContactos():
  print("Lista de contactos")
  Contador = 0
  print(f"{'Nombre':40}{'Teléfono':20}")
  for i in range(1,len(Contactos["Nombres"])+1,1):
    print(f"{Contactos["Nombres"][i-1]:40}{Contactos["Telefono"][i-1]:20}")
    Contador +=1

print("1. Agregar contacto\n2. Buscar contacto\n3. Lista de contactos\n4. Salir")
while True:
  Opcion = int(input("Escribe el númer de la opción: "))
  match Opcion:
    case 1:
      AgregarContacto()
    case 2:
      BuscarContacto()
    case 3:
      ListaContactos()
    case 4:
      break
    case _:
      print("Opción invalida")

```

**18. Contar vocales**

Crea una función que reciba una cadena y devuelva cuántas vocales tiene.

```python
VocalA = 0
VocalE = 0
VocalI = 0
VocalO = 0
VocalU = 0

Cadena = ""

def ContarVocales(Cadena):
  print(Cadena)
  VocalA = Cadena.lower().count('a')
  VocalE = Cadena.lower().count('e')
  VocalI = Cadena.lower().count('i')
  VocalO = Cadena.lower().count('o')
  VocalU = Cadena.lower().count('u')
  return VocalA, VocalE, VocalI, VocalO, VocalU

A, E, I, O, U = ContarVocales(input("Escribe una cadena de text: "))

print("Vocal A: ",A)
print("Vocal E: ",E)
print("Vocal I: ",I)
print("Vocal O: ",O)
print("Vocal U: ",U)
```

**21. Invertir número**

Dado un número entero, inviértelo usando un ciclo.

Ejemplo:
1234 → 4321

```python
NumeroOriginal = input("Dame el numero a invertir: ")
NumeroInvertido = ""

if NumeroOriginal.isdigit():
  for i in range(len(NumeroOriginal), 0, -1):
    NumeroInvertido = NumeroInvertido + NumeroOriginal[i-1]
else:
  print("La cadena no es un número")

print("Numero invertido: ", NumeroInvertido)
```
