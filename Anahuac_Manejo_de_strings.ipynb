**10. Iniciales**

Pide nombre completo y muestra las iniciales en mayúscula.
Ejemplo:
juan perez lopez → JPL

```python
Frase = input("Escribe tu nombre completo:")
Iniciales = ""
CantidadPalabras = len(Frase.split())
PalabrasSeparadas = Frase.split()

for i in range(0,CantidadPalabras,1):
  Iniciales = Iniciales + PalabrasSeparadas[i][0:1].upper()
print(Iniciales)
```

**18. Contar tipo de caracteres**

En una frase contar:

Cuántas letras

Cuántos números

Cuántos espacios

Cuántos símbolos

```python
Frase = "@Augusto Lorenzo% Gassós# &912"
Numeros = 0
Espacios = 0
Letras = 0
Simbolos = 0

for i in Frase:
  if i.isnumeric():
    Numeros += 1
  elif i.isspace():
    Espacios += 1
  elif i.isalpha():
    Letras += 1
  elif i.isascii():
    Simbolos += 1

print("Letras: ", Letras)
print("Números: ", Numeros)
print("Espacios: ", Espacios)
print("Símbolos: ", Simbolos)
```
