**1. Costo total con descuento e impuesto**

Una tienda vende un producto en $120.
Hay un descuento del 15% y luego se aplica un IVA del 21% sobre el precio con descuento.

Calcula el precio final a pagar usando una sola expresión.

```python
precio = (120-(120*.15))+((120-(120*.15))*.21)
print(precio)
```
