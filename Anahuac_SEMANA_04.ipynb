```python
#lista de secuencias de ADN
dna_sequence = ['GCT', 'AGC', 'AGG', 'TAA', 'ACT', 'CAT', 'TAT', 'CCC', 'ACG', 'GAA', 'ACC', 'GGA']

#elementos a buscar
i_to_find = 'ACG'

#variable boleana
i_found = False

#recorrer la lista
for i in dna_sequence:
  if i == i_to_find:
    i_found = True
    break # Salir del bucle una vez que se encuentra el elemento

#verificar resultado (fuera del bucle)
if i_found:
  print("i found")
else:
  print("i not found")
```
