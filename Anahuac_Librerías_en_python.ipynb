**2. Juego de adivinanza**

Genera un número aleatorio entre 1 y 100.

El usuario debe adivinarlo.

El programa indica si es mayor o menor.

Cuenta los intentos.

Usa: random.randint()

```python
import random as rdm

Contador = 0
NumeroAleatorio = rdm.randint(1, 100)

while True:
  NumeroUsuario = int(input("Dame un número entero: "))
  Contador +=1
  if NumeroAleatorio == NumeroUsuario:
    break
  else:
    if NumeroUsuario < NumeroAleatorio:
      print("El número que escibiste es menor.")
    else:
      print("El número que escibiste es mayor.")
    print("No has adivinado el número\nIntenta nuevamente ....\n\n")

print(f"Te tomó {Contador} intentos adivivar el número")
```

**3. Pide la fecha de nacimiento del usuario y calcula:**

Edad en años

Días totales vividos

```python
from IPython.utils.io import atexit
import datetime
from dateutil.relativedelta import relativedelta

FechaNacimiento = input("Escribe tu fecha de nacimiento: ")
FechaNacimiento = datetime.datetime.strptime(FechaNacimiento, "%d/%m/%Y")

Edad = relativedelta(datetime.datetime.now(), FechaNacimiento)

Fecha1 = datetime.datetime(int(FechaNacimiento.strftime("%Y")),int(FechaNacimiento.strftime("%m")),int(FechaNacimiento.strftime("%d")))
Fecha2 = datetime.datetime(int(datetime.datetime.now().strftime("%Y")),int(datetime.datetime.now().strftime("%m")),int(datetime.datetime.now().strftime("%d")))

print("Años: ", Edad.years)
print("Días: ", Fecha2-Fecha1)
```

**8. Validador de email**

Crea una función que valide si un string es un email válido.

Usa: re.match() con una expresión regular.

```python
import re

ExpresionCorreo = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')
if re.match(ExpresionCorreo, "augusto.lorenzo.acero@gmail.com"):
  print("Correo válido")
else:
  print("Correo invalido")

```
