```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(range(-10,10))

plt.figure()

# linear
plt.subplot(221)
plt.plot(x, x)
plt.title("linear")

# log
plt.subplot(223)
plt.plot(x, np.log(x))
plt.title("log")

# cuadrado
plt.subplot(224)
plt.plot(x, x**2)
plt.title("pow 2")

# cubo
plt.subplot(222)
plt.plot(x, x**3)
plt.title("pow 3")
plt.grid(True)

plt.show()
```
