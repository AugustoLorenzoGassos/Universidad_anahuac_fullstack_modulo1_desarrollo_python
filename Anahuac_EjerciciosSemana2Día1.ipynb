Ejercicio 1

```python
def Suma(A, B):
  Resultado = A + B
  return Resultado

def Par(x):
  if x%2 == 0:
    return True
  else:
    return False

def SumaTres(A, B, C):
  return A+B+C

print("El resultado de sumar dos numeros: ",Suma(4,5))
print("El resultado para saber si el número es par o no: ",Par(Suma(4,5)))
print("El resultado de sumar 3 números y sabes si es par o no: ",SumaTres(10,20,Suma(4,5)),Par(SumaTres(10,20,Suma(4,5))))

ListaNumeros = []
for i in range(1,9,1):
  ListaNumeros.append(i)
print("Lista de números del 1 al 8", ListaNumeros)

```

Ejercicio 2

```python
ListaNumeros = []

def LlenarLista():
  for i in range(1,9,1):
    ListaNumeros.append(i)
  print("Lista creada: ",ListaNumeros)

def RevisarPar():
  for i in ListaNumeros:
    if Par(i):
      print(f"El numero {i} es par")
    else:
      print(f"El numero {i} no es par")

def AgregarFinal():
  ListaNumeros.append(9)
  print(ListaNumeros)

LlenarLista()
RevisarPar()
AgregarFinal()



```

Ejercicio 3

```python
while True:
  NumeroA = int(input("Escribe el número inicial de la secuencia: "))
  NumeroB = int(input("Escribe el número fila de la secuencia (Debe ser mayor o igual que el número inicial ): "))
  if NumeroA <=  NumeroB:
    break
  else:
    print("Intenta nuevamente")

Inicio = NumeroA
CadenaNumeros = ""

while Inicio <= NumeroB:
  print(Inicio, end=",")
  if Inicio < NumeroB:
    CadenaNumeros = CadenaNumeros + str(Inicio)+","
  else:
    CadenaNumeros = CadenaNumeros + str(Inicio)
  Inicio +=1

print("\n")
print(CadenaNumeros)

```
